# Section 5: Data Science
Question: Is there a correlation with ratings and worldwide box office?

I chose this question because it helps reveal whether a movie’s rating actually predicts how much money it earns at the box office. Instead of just looking at the correlation, I want to use a machine learning model to see whether ratings can predict the box office of a movie. Instead of just looking at the correlation, I used a simple machine learning model to see whether ratings can predict the worldwide box office of a movie. First, I had to cleaned the "Rating" column because ratings appeared as strings like "6.7/10", so I converted each one into just the numeric score by splitting the string and taking the part before the “/10”. After converting the cleaned values to floats, I filtered out rows where the rating was missing so the model wouldn’t try to learn from data that showed up as NaN. Next, I used a Linear Regression model with Rating as the independent variable and Worldwide box office as the dependent variable to find a straight-line relationship between how high a movie is rated and how much money it earns. After fitting the model, I generated a new column called "Predicted Worldwide", which shows the estimated revenue based only on the rating. Then I made a smaller DataFrame that contained the columns "Rating", "$Worldwide", and "Predicted Worldwide" so it was easier to read the relevenat data. Even though rating alone can’t explain every factor that affects box office performance, the model still shows a strong positive linear trend that higher-rated movies are generally predicted to earn more as shown the correlation coefficient calculated below which is 0.99999. 

```python
from sklearn.linear_model import LinearRegression

df["Rating"] = df["Rating"].str.replace("/10", "").astype(float)

df = df[df["Rating"] > 0]

ind = ["Rating"]
dep = "$Worldwide"
model = LinearRegression()

model.fit(df[ind], df[dep])

df["Predicted Worldwide"] = model.predict(df[ind])

df_small = df[["Rating", "$Worldwide", "Predicted Worldwide"]]
df_small
```

```python
r_model = df_small["Rating"].corr(df_small["Predicted Worldwide"])
r_model
```

# Section 6: A Different Data Visualization
The residual plot helps reveal how well a linear model can predict worldwide box office using only movie ratings. Instead of simply looking at the trend, the residual plot shows the errors the model makes for each prediction. The residual plot shows as ratings increase, the spread of residuals becomes extremely wide. This indicates that the linear model does not fit the data well. This shows that the linear model is not capturing the true relationship between rating and worldwide box office. This means that a movie’s rating does not strongly predict how much money it makes at the worldwide box office. Movies with similar ratings can earn very different amounts of money, and many low rated movies still make a lot. The residual plot reveals that the relationship is nonlinear and highly variable, meaning other factors like genre or language probably play a larger role in box office success than rating alone. The correlation coefficient between rating and worldwide box office was also only 0.1908, which reflects a very weak linear relationship. 

```python
df_small["Residual"] = df_small["$Worldwide"] - df_small["Predicted Worldwide"]
```

```python
df_small.plot.scatter(x="Rating", y = "Residual", title= "Residuals of Linear Model Predicting Worldwide Revenue from Rating", xlabel = "Rating (out of 10)", ylabel = "Residual")
```

```python
r_actual = df_small["Rating"].corr(df_small["$Worldwide"])
r_actual
```
